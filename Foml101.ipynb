```python
import pandas as pd

df = pd.read_csv("pit_stops.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
total_pit_stops = len(df)
total_pit_stops
```

```python
num_races = df["raceId"].nunique()
num_drivers = df["driverId"].nunique()

num_races, num_drivers
```

```python
avg_pitstops_per_race = total_pit_stops / num_races
avg_pitstops_per_race
```

```python
df.isnull().sum()
```

```python
df["duration_sec"] = df["milliseconds"] / 1000
df["duration_sec"].describe()
```

```python
long_stops = df[df["duration_sec"] > 30]
long_stops.shape
```

```python
long_stops[["driverId", "raceId", "lap", "duration_sec"]].head()
```

```python
import matplotlib.pyplot as plt

plt.hist(df["duration_sec"], bins=5)
plt.xlabel("Pit Stop Duration (seconds)")
plt.ylabel("Frequency")
plt.title("Distribution of Pit Stop Durations")
plt.show()
```

```python
plt.boxplot(df["duration_sec"], vert=False)
plt.xlabel("Pit Stop Duration (seconds)")
plt.title("Pit Stop Duration Outliers")
plt.show()
```
